Load the data. Notice that we are splitting the data into training, development, and test. We also have a small subset of the training data called mini_train_data and mini_train_labels that you should use in all the experiments below, unless otherwise noted.

```python
# Load the digit data either from mldata.org, or once downloaded to data_home, from disk. The data is about 53MB so this cell
# should take a while the first time your run it.
mnist = fetch_mldata('MNIST original', data_home='~/datasets/mnist')
X, Y = mnist.data, mnist.target

# Rescale grayscale values to [0,1].
X = X / 255.0

# Shuffle the input: create a random permutation of the integers between 0 and the number of data points and apply this
# permutation to X and Y.
# NOTE: Each time you run this cell, you'll re-shuffle the data, resulting in a different ordering.
shuffle = np.random.permutation(np.arange(X.shape[0]))
X, Y = X[shuffle], Y[shuffle]

print('data shape: ', X.shape)
print('label shape:', Y.shape)

# Set some variables to hold test, dev, and training data.
test_data, test_labels = X[61000:], Y[61000:]
dev_data, dev_labels = X[60000:61000], Y[60000:61000]
train_data, train_labels = X[:60000], Y[:60000]
mini_train_data, mini_train_labels = X[:1000], Y[:1000]
```

(1) Create a 10x10 grid to visualize 10 examples of each digit. Python hints:

- plt.rc() for setting the colormap, for example to black and white
- plt.subplot() for creating subplots
- plt.imshow() for rendering a matrix

```python
def P1(num_examples=10): 
    # initialize empty matrix of 100 examples, each with 784 data points
    data = np.empty((100, 784))
    fig, grid = plt.subplots(10, num_examples, sharex=True, sharey=True)
    for digit in range(0,10):
        # populate data matrix from training data
        data[num_examples*digit:num_examples*digit+num_examples,] = mini_train_data[np.where(mini_train_labels==digit)][0:10,]
        # plot ten examples for each digit
        for example in range(0, num_examples):
            matrix = data[digit*num_examples + example].reshape(28, 28)
            grid[digit][example].imshow(matrix, cmap='Greys',  interpolation='nearest')
            grid[digit][example].axis('off')
P1(10)
```

(2) Evaluate a K-Nearest-Neighbors model with k = [1,3,5,7,9] using the mini training set. Report accuracy on the dev set. For k=1, show precision, recall, and F1 for each label. Which is the most difficult digit?

- KNeighborsClassifier() for fitting and predicting
- classification_report() for producing precision, recall, F1 results

```python
def P2(k_values):
    # We will return accuracy in a list for each k in input list
    accuracy_list = []
    for k in k_values:
        # train model
        model = KNeighborsClassifier(n_neighbors=k)
        model.fit(mini_train_data, mini_train_labels)
        # test model on dev data and evaluate accuracy
        dev_predicted_labels = model.predict(dev_data)
        wrong_prediction = (dev_predicted_labels != dev_labels)
        accuracy = 1 - np.sum(wrong_prediction)/np.size(dev_labels)
        accuracy_list.append(accuracy)
        # for k=1, give additional information
        if k == 1:
            for digit in range(10):
                # precision = how many selected items are relevant
                precision = np.sum(np.logical_and(dev_predicted_labels==dev_labels, dev_labels==digit))/np.sum(dev_predicted_labels == digit)
                # recall = how many relevant items are selected
                recall = np.sum(np.logical_and(dev_predicted_labels==dev_labels, dev_labels==digit))/np.sum(dev_labels == digit)
                F1 = 2 * (precision * recall) / (precision + recall)
                print('For digit = ' + str(digit) + ': precision = ' + "{:.2f}".format(precision) +
                      ', recall = ' + "{:.2f}".format(recall) + ', F1 = ' + "{:.2f}".format(F1))
    return accuracy_list

k_values = [1, 3, 5, 7, 9]
P2(k_values)
```

ANSWER:  
9 is the most difficult, using F1 as the metric, presumably in part because it resembles a 4.

(3) Using k=1, report dev set accuracy for the training set sizes below. Also, measure the amount of time needed for prediction with each training size.

- time.time() gives a wall clock value you can use for timing operations

```python
def P3(train_sizes, accuracies):
    times = []
    for size in train_sizes:
        # train model
        model = KNeighborsClassifier(n_neighbors=1)
        model.fit(train_data[:size], train_labels[:size])
        # test model on dev data and evaluate accuracy
        time1 = time.time()
        dev_predicted_labels = model.predict(dev_data)
        prediction_time = time.time() - time1
        wrong_prediction = (dev_predicted_labels != dev_labels)
        accuracy = 1 - np.sum(wrong_prediction)/np.size(dev_labels)
        accuracies.append(accuracy)
        times.append(prediction_time)
    return accuracies, times

train_sizes = [100, 200, 400, 800, 1600, 3200, 6400, 12800, 25000]
accuracies = []
accuracies, times = P3(train_sizes, accuracies)
```

It is, however, clear from the scatterplot residuals that the model is overestimating accuracy at large and small sizes. The residuals are not randomly distributed, which suggests that we have violated an OLS assumption and do not have an appropriate model for precise predictions, particularly for extrapolation. (6) A common image processing technique is to smooth an image by blurring. The idea is that the value of a particular pixel is estimated as the weighted combination of the original value and the values around it. Typically, the blurring is Gaussian -- that is, the weight of a pixel's influence is determined by a Gaussian function over the distance to the relevant pixel.

Implement a simplified Gaussian blur by just using the 8 neighboring pixels: the smoothed value of a pixel is a weighted combination of the original value and the 8 neighboring values. Try applying your blur filter in 3 ways:
- preprocess the training data but not the dev data
- preprocess the dev data but not the training data
- preprocess both training and dev data

Note that there are Guassian blur filters available, for example in scipy.ndimage.filters. You're welcome to experiment with those, but you are likely to get the best results with the simplified version I described above.

```python
def P6():
    blurred_train = blur(train_data)
    blurred_dev = blur(dev_data)
    
    # print sample of images pre- and post-blur
    sample = 1
    print_sample(train_data, sample)
    print_sample(blurred_train, sample)
    
    # -- train and evaluate baseline and three proposed scenarios --
    # baseline
    acc_base = train_and_eval(train_data, train_labels, dev_data, dev_labels)
    print(acc_base)
    # preprocess the training data but not the dev data
    acc1 = train_and_eval(blurred_train, train_labels, dev_data, dev_labels)
    # preprocess the dev data but not the training data
    acc2 = train_and_eval(train_data, train_labels, blurred_dev, dev_labels)
    # preprocess both training and dev data
    acc3 = train_and_eval(blurred_train, train_labels, blurred_dev, dev_labels)
    print(acc1, acc2, acc3)
    
def train_and_eval(train_data, train_labels, test_data, test_labels):
    model = KNeighborsClassifier(n_neighbors=1)
    model.fit(train_data, train_labels)
    test_predicted_labels = model.predict(test_data)
    wrong_prediction = (test_predicted_labels != test_labels)
    accuracy = 1 - np.sum(wrong_prediction)/np.size(test_labels)
    return accuracy
    
def blur(data):
    # blurs a set of images
    
    # copy original dataset mostly to get get an pre-existing format to insert blurred data into
    # deep copy so that working on blurred data does not corrupt original dataset
    blurred_data = deepcopy(data)
    
    # for each image in dataset
    for i in range(np.shape(data)[0]):
        image = data[i]
        # reshape for intuitive understanding of what to blur
        image = image.reshape(28, 28)
        # deep copy image so that new image can be changed without affecting old image
        new_image = deepcopy(image)

        for row in range(28):
            for col in range(28):
                # my algorithm tries find all neighboring pixels (ignoring those that would be outside the image frame),
                # then counts how many edges of the image frame the pixel touches, with which it can determine number of
                # neighboring pixels, and finally calculates a new value for the pixel.

                # initialize total value at value of pixel in question
                total = float(image[row][col])

                
                # enumerate a list of tuples that specify neighboring pixels
                neighbors = [(row-1,col-1), (row-1,col), (row-1,col+1), 
                             (row,  col-1),              (row,  col+1), 
                             (row+1,col-1), (row+1,col), (row+1,col+1)]
                for pixel in neighbors:
                    # if try fails, it means the pixel doesn't exist because it is beyond an edge of the frame
                    try:
                        # use old image for values, since new image is updating as it goes
                        total += float(image[pixel[0],pixel[1]])
                    except:
                        pass
                    
                # how many edges does the pixel touch? (1 for edge of frame, 2 for corner pixels) 
                edges = 0
                if row == 0 or row == 27:
                    edges += 1
                if col == 0 or col == 27:
                    edges += 1
                # get total neighboring pixel count based on number of edges
                if edges == 0:
                    pixel_count = 9
                elif edges == 1:
                    pixel_count = 6
                else: # edges == 2
                    pixel_count = 4

                # set new pixel value based on total from surrounding pixels divided by total count of pixels
                new_image[row][col] = total / pixel_count
                
        # reshape and store
        new_image = new_image.reshape(784,)
        blurred_data[i] = new_image
        
    return blurred_data

def print_sample(data, sample=1):
    for num in range(sample):
        image = data[num]
        matrix = image.reshape(28, 28)
        plt.figure(figsize=(1.5, 1.5))
        plt.axis('off')
        plt.imshow(matrix, cmap='Greys',  interpolation='nearest')
    return

P6()
```

(7) Fit a Naive Bayes classifier and report accuracy on the dev data. Remember that Naive Bayes estimates P(feature|label). While sklearn can handle real-valued features, let's start by mapping the pixel values to either 0 or 1. You can do this as a preprocessing step, or with the binarize argument. With binary-valued features, you can use BernoulliNB. Next try mapping the pixel values to 0, 1, or 2, representing white, grey, or black. This mapping requires MultinomialNB. Does the multi-class version improve the results? Why or why not?

```python
def P7():

    # I will start with a threshold of 0, which will round up to 1 if the pixel has any level of fill at all.
    # A reasonable argument could be made other values, but I chose this as it clearly distinguishes where there the
    # writing instrument touched.
    b_train = binarize(train_data, threshold=0)
    b_dev = binarize(dev_data, threshold=0)

    # model and evaluate
    nb = BernoulliNB()
    nb.fit(b_train, train_labels)
    print('binary accuracy of:', nb.score(b_dev, dev_labels))

    # superimposing two binarize commands can let us make three thresholds. These roughly indicate white, grey, and black.
    b_train2 = binarize(train_data, threshold=0) + binarize(train_data, threshold=.5)
    b_dev2 = binarize(dev_data, threshold=0) + binarize(dev_data, threshold=.5)

    nb2 = MultinomialNB()
    nb2.fit(b_train2, train_labels)
    print('multinomial accuracy of:', nb2.score(b_dev2, dev_labels))

P7()
```

ANSWER:  
My hypothesis was that multinomial would not show a significant improvement over binomial and may actually perform worse due primarily to overfitting. My reasoning was that it is already a strong jump to say that pixels contribute to the image classification independent of their neighbors. For example, a 1 drawn on the left side of the frame versus the right side will have totally different pixel matrices. There is a lot of information that cannot be captured by Naive Bayes, at least in this formulation, so I would only assume a very rough accuracy, as we indeed saw. Furthermore, breaking the pixels down into grey and black doesn't necessarily add any meaningful information. One person may just write with lighter pencil strokes than another. My assumption was that we would be overfitting by adding additional states that didn't capture much more meaningful information. One place where they could potentially contribute some information could be, for example, at the top or bottom of a number 1. A person is likely to apply less pressure at the start and/or end of a stroke, so I could see some marginal informational capture there, but in general this will not be adding much value at the cost of overfitting.  

This does seem to be the case, as binary features showed a very rough accuracy (.826) and multinomial features showed a slight dropoff (.822). Both of these represent a substantial improvement over a hypothetical baseline of random guessing (.100), so this methodology is somewhat useful. Further feature engineering by aggregating pixels in some meaningful way could likely show improvement.

(11) Remember that a strongly calibrated classifier is rougly 90% accurate when the posterior probability of the predicted class is 0.9. A weakly calibrated classifier is more accurate when the posterior is 90% than when it is 80%. A poorly calibrated classifier has no positive correlation between posterior and accuracy.

Train a BernoulliNB model with a reasonable alpha value. For each posterior bucket (think of a bin in a histogram), you want to estimate the classifier's accuracy. So for each prediction, find the bucket the maximum posterior belongs to and update the "correct" and "total" counters.

How would you characterize the calibration for the Naive Bayes model?

```python
def P11(buckets, correct, total):
    # fit model
    nb = BernoulliNB(alpha=0)
    nb.fit(train_data, train_labels)
    # for each image one at a time
    for i in range(np.shape(dev_data)[0]):
        image = dev_data[i]
        # find the max probability and corresponding digit
        prob_vect = nb.predict_proba(image.reshape(1, -1))
        prob = np.amax(prob_vect)
        digit = np.argmax(prob_vect)
        # update correct and total accordingly
        for index in range(len(buckets)):
            threshold = buckets[index]
            if prob > threshold:
                continue
            else:
                total[index] += 1
                if digit == dev_labels[i]:
                    correct[index] += 1
                break
    return

# initialize histogram buckets
buckets = [0.5, 0.9, 0.999, 0.99999, 0.9999999, 0.999999999, 0.99999999999, 0.9999999999999, 1.0]
correct = [0 for i in buckets]
total = [0 for i in buckets]

P11(buckets, correct, total)

# print results
for i in range(len(buckets)):
    accuracy = 0.0
    if (total[i] > 0):
        accuracy = correct[i] / total[i]
    print('p(pred) <= %.13f    total = %3d    accuracy = %.3f' %(buckets[i], total[i], accuracy))
```

ANSWER:  
The classifier is weakly calibrated as accuracy is trending up as probability increases. It is noteworthy that this example confirms that Naive Bayes tends to suffer from overconfidence. Many extremely high probability values yield fairly low accuracy in their predictions.
